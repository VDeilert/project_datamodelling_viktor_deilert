## Visar klasser och dess kurs, lärare samt om läraren är anställd på skolan eller konsult

```python
%%sql
SELECT 
    kl.namn AS Klass,
    ku.namn AS Kurs,
    u.fornamn || ' ' || u.efternamn AS Utbildare,
    CASE 
        WHEN ko.utbildare_id IS NOT NULL THEN 'Konsult'
        ELSE 'Anställd'
    END AS Status
FROM Klass_Kurs kk
JOIN Klass kl ON kk.klass_id = kl.klass_id
JOIN Kurs ku ON kk.kurs_id = ku.kurs_id
JOIN Utbildare u ON kk.utbildare_id = u.utbildare_id
LEFT JOIN Konsult ko ON u.utbildare_id = ko.utbildare_id;
```

## Antalet studenter per klass

```python
%%sql
SELECT 
    k.namn AS Klassnamn, 
    COUNT(s.student_id) AS Antal_Studenter
FROM Klass k
LEFT JOIN Student s ON k.klass_id = s.klass_id
GROUP BY k.namn;
```

## Visar konsulter, företaget dem arbetar för, deras timtaxa samt företagets adress

```python
%%sql
SELECT 
    u.fornamn || ' ' || u.efternamn AS Konsultnamn,
    f.namn AS Arbetar_För,
    ko.timarvode_sek AS Timtaxa,
    f.adress AS Företagsadress
FROM Utbildare u
JOIN Konsult ko ON u.utbildare_id = ko.utbildare_id
JOIN Foretag f ON ko.foretag_id = f.foretag_id;
```
